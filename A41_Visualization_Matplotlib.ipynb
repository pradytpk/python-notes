```python
import matplotlib.pyplot as plt
```

## Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("this is my x axis")
plt.ylabel("this is my x axis")
```

```python
plt.scatter(x,y)
plt.xlabel("this is my x axis")
plt.ylabel("this is my x axis")
plt.title("X vs Y")
```

```python
plt.scatter(x,y,c='r')
plt.xlabel("this is my x axis")
plt.ylabel("this is my x axis")
plt.title("X vs Y")
```

```python
plt.scatter(x,y,c='#2AEE98',alpha=0.9)
plt.xlabel("this is my x axis")
plt.ylabel("this is my x axis")
plt.title("X vs Y")
```

```python
plt.scatter(x,y,c='#2AEE98',alpha=0.9,marker='*')
plt.xlabel("this is my x axis")
plt.ylabel("this is my x axis")
plt.title("X vs Y")
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='r',alpha=0.9,marker='*')
plt.xlabel("this is my x axis")
plt.ylabel("this is my x axis")
plt.title("X vs Y")
plt.grid()
```

```python
x=[3,4,5,6,7,8]
y=[4,5,6,7,8,9]
plt.figure(figsize=(6,2))
plt.scatter(x,y)
```

## Line Plot

```python
x=np.linspace(1,10,100)
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y,'--r')
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sin wave")
plt.grid()
```

```python
x=[3,4,5,6,7,8]
y=[4,5,6,7,8,9]
plt.figure(figsize=(6,2))
plt.plot(x,y)
```

## Bar Plot

```python
x=['a','b','c','d','e']
y=np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.bar(x,y)
```

```python
plt.bar(x,y,color="orange")
plt.xlabel("Representing my categorical data")
plt.ylabel("Representing my numerical data")
plt.title("bar plot")
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color="orange")
plt.xlabel("Representing my categorical data")
plt.ylabel("Representing my numerical data")
plt.title("bar plot")
plt.grid()
```

### Bar Horizontal

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color="orange")
plt.xlabel("Representing my categorical data")
plt.ylabel("Representing my numerical data")
plt.title("bar plot")
plt.grid()
```

## Histogram

```python
data =[1,2,3,4,5,6,6,6,7,8,9,9,9,10,1,0,1,1,1]
```

```python
plt.hist(data)
```

```python
plt.figure(figsize=(6,3))
plt.hist(data,color='pink')
plt.xlabel("Representing unique Value")
plt.ylabel("Representing frequency")
plt.title("Histogram plot")
plt.grid()
plt.savefig("test.png")
```

## 3D Graph

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='r')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
