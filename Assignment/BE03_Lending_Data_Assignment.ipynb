## Replace '.' with column mean

```python
lend.dtypes
```

```python
# Check for the presence of '.' in the entire DataFrame
dot_present = lend.applymap(lambda x: str(x) =='    .  ')
dot_present.sum()
```

```python
for column in lend.columns:
    if dot_present[column].any():
        column_mean = pd.to_numeric(lend[column], errors='coerce').mean()
        lend[column] = lend[column].replace('    .  ', column_mean)
```

```python
dot_present = lend.applymap(lambda x: str(x) =='    .  ')
dot_present.sum()
```
