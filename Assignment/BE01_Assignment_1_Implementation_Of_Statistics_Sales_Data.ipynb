## Trend analysis of sales over time (monthly, quarterly, annually).

```python
sales_data['Date'] = pd.to_datetime(sales_data['Year'].astype(str) + '-' + sales_data['Month Name'], format='%Y-%B')

monthly_sales = sales_data.groupby('Date')['Sales'].sum().reset_index()

monthly_sales['Quarter'] = monthly_sales['Date'].dt.to_period('Q')
quarterly_sales = monthly_sales.groupby('Quarter')['Sales'].sum().reset_index()

monthly_sales['Year'] = monthly_sales['Date'].dt.year
annual_sales = monthly_sales.groupby('Year')['Sales'].sum().reset_index()

quarterly_sales['Quarter'] = quarterly_sales['Quarter'].astype(str)
```

```python
fig1 = px.line(monthly_sales, x='Date', y='Sales', title='Monthly Sales Trend')
fig1.show()
```

```python
fig2 = px.line(quarterly_sales, x='Quarter', y='Sales', title='Quarterly Sales Trend')
fig2.show()
```

```python
fig3 = px.line(annual_sales, x='Year', y='Sales', title='Annual Sales Trend')
fig3.show()
```

# Market Analysis

## Market share analysis by country or region.

```python
country_sales = sales_data.groupby('Country')['Units Sold'].sum().reset_index()

total_sales = country_sales['Units Sold'].sum()
country_sales['Market Share'] = (country_sales['Units Sold'] / total_sales) * 100

fig = px.pie(country_sales, names='Country', values='Market Share',
             title='Market Share by Country')

fig.show()
```

# Discount Impact Analysis

## Impact of discounts on sales volume and profit.

```python
sales_data['Discounted Sales'] = sales_data['Gross Sales'] * (1 - sales_data['Discounts'])
sales_data['Discounted Profit'] = sales_data['Discounted Sales'] - sales_data['COGS']


fig = px.scatter(sales_data, x='Discounts', y=['Discounted Sales', 'Discounted Profit'],
                 labels={'variable': 'Metrics'},
                 title='Impact of Discounts on Sales Volume and Profit')

fig.show()
```

## Correlation between discount levels and sales/profit increases.

```python
# Calculate the increase in Sales and Profit due to discounts
sales_data['Increase in Sales'] = sales_data['Discounted Sales'] - sales_data['Sales']
sales_data['Increase in Profit'] = sales_data['Discounted Profit'] - sales_data['Profit']

# Calculate the correlation coefficients
correlation_sales = np.corrcoef(sales_data['Discounts'], sales_data['Increase in Sales'])[0, 1]
correlation_profit = np.corrcoef(sales_data['Discounts'], sales_data['Increase in Profit'])[0, 1]

fig = px.scatter(sales_data, x='Discounts', y=['Increase in Sales', 'Increase in Profit'],
                 labels={'variable': 'Metrics'},
                 title=f'Correlation between Discounts and Sales/Profit Increases <br>'
                       f'Correlation with Sales: {correlation_sales:.2f}, '
                       f'Correlation with Profit: {correlation_profit:.2f}')

fig.show()
```

## Effectiveness of discount strategies over time.

```python
sales_data['Discounted Sales'] = sales_data['Gross Sales'] * (1 - sales_data['Discounts'])
sales_data['Discounted Profit'] = sales_data['Discounted Sales'] - sales_data['COGS']

fig = px.line(sales_data, x='Month Name', y=['Discounted Sales', 'Discounted Profit'],
              labels={'variable': 'Metrics'},
              title='Effectiveness of Discount Strategies Over Time')

fig.show()
```
