## Task 3: Pie Chart for Number of Models by Manufacturer

```python
# Assuming you have a DataFrame named 'car_data'
# You can group the data by manufacturer and count the number of models
manufacturer_counts = car_data['name'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(manufacturer_counts, labels=manufacturer_counts.index, autopct='%1.1f%%')
plt.title('Number of Models by Manufacturer')
plt.show()

# Find the manufacturer with the largest releases
largest_manufacturer = manufacturer_counts.idxmax()
print(f"The manufacturer with the largest releases is: {largest_manufacturer}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Task 5.1: Draw a Simple plot
X = [1, 2, 3, 4]
y = [20, 21, 20.5, 20.8]
plt.figure(figsize=(8, 4))
plt.plot(X, y)
plt.title('Simple Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Task 5.2: Configure the line and markers in a simple plot
plt.figure(figsize=(8, 4))
plt.plot(X, y, linestyle='--', marker='o', color='b')
plt.title('Line and Markers Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Task 5.3: Configure the axes
plt.figure(figsize=(8, 4))
plt.plot(X, y, linestyle='-', marker='o', color='g')
plt.title('Configured Axes Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0, 5)
plt.ylim(19, 22)
plt.show()
```

```python
# Task 5.4: Give the title of the graph and labels of the x-axis and y-axis
plt.figure(figsize=(8, 4))
plt.plot(X, y, linestyle='-', marker='o', color='r')
plt.title('Customized Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

```python
# Task 5.5: Give error bar if y_error = [0.12, 0.13, 0.2, 0.1]
y_error = [0.12, 0.13, 0.2, 0.1]
plt.figure(figsize=(8, 4))
plt.errorbar(X, y, yerr=y_error, linestyle='-', marker='o', color='purple')
plt.title('Error Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Task 5.6: Define width, height, and DPI
plt.figure(figsize=(4, 5), dpi=100)
plt.plot(X, y, linestyle='-', marker='o', color='orange')
plt.title('Custom Size and DPI Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Task 5.7: Give a font size of 14
plt.figure(figsize=(8, 4))
plt.plot(X, y, linestyle='-', marker='o', color='b')
plt.title('Custom Font Size Plot', fontsize=14)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Task 5.8: Draw a scatter graph of 50 random values of x and y-axis
x_random = np.random.rand(50)
y_random = np.random.rand(50)
plt.figure(figsize=(8, 4))
plt.scatter(x_random, y_random, marker='o', color='green')
plt.title('Scatter Plot of Random Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Task 5.9: Create a DataFrame from the given data
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'female': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}
```

```python
df = pd.DataFrame(data)

# Scatterplot of preTestScore and postTestScore with size determined by age
plt.figure(figsize=(8, 4))
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'], alpha=0.5)
plt.title('Scatter Plot with Size by Age')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

```python
# Task 5.10: Scatterplot with size = 300 and color determined by sex
colors = df['female'].map({0: 'blue', 1: 'pink'})
plt.figure(figsize=(8, 4))
plt.scatter(df['preTestScore'], df['postTestScore'], s=300, c=colors, alpha=0.7)
plt.title('Scatter Plot with Custom Size and Color')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

## Case Study

```python
# Read CSV files
df_math_scores = pd.read_csv("C:\Study\Data Science\Datasets\MathScoreTerm1.csv")
df_ds_score = pd.read_csv("C:\Study\Data Science\Datasets\DSScoreTerm1.csv")
df_physics_score = pd.read_csv("C:\Study\Data Science\Datasets\PhysicsScoreTerm1.csv")
```

```python
# Remove columns
del df_math_scores["Name"]
del df_math_scores["Ethinicity"]

del df_ds_score["Name"]
del df_ds_score["Ethinicity"]

del df_physics_score["Name"]
del df_physics_score["Ethinicity"]
```

```python
# Fill missing values with zeros
df_math_scores.fillna(0, inplace=True)
df_ds_score.fillna(0, inplace=True)
df_physics_score.fillna(0, inplace=True)
```

```python
# Merge DataFrames
merged_df = df_math_scores.merge(df_ds_score, on="ID", suffixes=('_math', '_ds')).merge(df_physics_score, on="ID", suffixes=('_ds', '_phys'))
```

```python
# Update Sex column
sex_mapping = {"M": 1, "F": 2}
merged_df["Sex"] = merged_df["Sex"].map(sex_mapping)
```

```python
# Save cleaned data
merged_df.to_csv("ScoreFinal.csv")
```
