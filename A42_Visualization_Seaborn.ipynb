# datasets

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("tips")
print(data.head())
```

```python
df = pd.DataFrame(np.random.rand(10,4), columns = ['a','b','c','d'])
```

```python
df.plot.bar()
```

```python
df.plot.barh(stacked = True)
```

# relplot

```python
sns.relplot(data= data, x= 'total_bill', y='tip', kind='scatter', col = 'sex', row= 'smoker', hue='time', size='day', style = 'size')
```

# scatterplot

```python
sns.scatterplot(data= data,x= 'total_bill', y='tip' )
```

```python
sns.lineplot(data= data,x= 'total_bill', y='tip' )
```

# distplot

```python
data.head()
```

```python
sns.displot(data['tip'], kde= True, color = 'r')
```

# barplot

```python
import numpy as np
```

```python
sns.barplot(data = data,  x= 'day', y = 'tip', estimator= np.mean)
```

# countplot

```python
sns.countplot(data= data, x='sex', hue= 'time')
```

# boxenplot

```python
df.plot.box()
```

```python
userspecific = dict(boxes = 'DarkGreen', whiskers = 'DarkOrange', medians = 'DarkBlue', caps = 'Gray')
```

```python
df.plot.box(color = userspecific)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.boxenplot(data= data, x= 'sex', y ='total_bill', hue = 'time')
plt.title("title")
plt.show()
```

# violin

```python
sns.violinplot(data= data, x= 'sex', y ='total_bill', hue= 'time')
```

# jointplot

```python
sns.jointplot(data= data ,   x= 'total_bill', y ='tip', color = 'gray', kind= 'kde')
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(data= data.corr(numeric_only=True), annot= True, cmap= 'Dark2')
```

```python
data.corr(numeric_only=True)
```
